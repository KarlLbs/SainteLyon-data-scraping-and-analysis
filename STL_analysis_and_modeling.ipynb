## Feature engineering

The data is already quite clean after the first step, so the only things we'll need to do are :


1- transform all times from hh:mm:ss strings to minutes :

```python
checkpoints = ["Pt0 time", "Pt1 time", "Pt2 time", "Pt3 time", "Pt4 time", "Pt5 time", "Pt6 time"]
for checkpoint in checkpoints:
    df[checkpoint] = df[checkpoint].apply(lambda x: sum(int(t) * sec for t, sec in zip(x.split(":"), [60, 1, 1/60])) if pd.notna(x) else np.nan)
```

2- Convert old categories names to the new ones :

```python
df['Category'] = df['Category'].replace({'SE F':'SE-M0F',
                                     'SE H':'SE-M0H',
                                     'V1 F':'M1-M2F',
                                     'V1 H':'M1-M2H',
                                     'V2 F':'M3-M4F',
                                     'V2 H':'M3-M4H',
                                     'V3 F':'M5-M6F',
                                     'V3 H':'M5-M6H',
                                     'V4 F':'M9-M10F', # Mapping the old 69-129yo category to the new 79-129 one instead of the 69-79 one
                                     'V4 H':'M9-M10H'})
pd.unique(df['Category'])
```

Now, let's not forget that every year has a different course, and consequently a different distance and elevation ! For now we'll take this into account with the simple assumption that 100m of elevation gain is equivalent to 1km flat. We'll calculate an effective distance based on that, and an adjusted finish times by normalizing results (as if every year's race was the same distance and elevation).

$effectiveDistance = realDistance + elevationGain/100$

```python
### Adjusted finish times distributions by year

# New dataframes from this part of the analysis to keep the original ones clean
df2 = df.copy()
cpdf = checkpoints_df.copy()

# Calculate the effective distance for each year
cpdf['Effective distance'] = cpdf['Pt6 distance'] + 0.01 * cpdf['Pt6 total elevation']

# Calculate the scaling factor relative to a reference year (2024)
reference_effective_distance = cpdf.loc[cpdf['Year'] == 2024, 'Effective distance'].values[0]
cpdf['Scaling factor'] = reference_effective_distance / cpdf['Effective distance']

# Adjust the finish times
df2 = df2.merge(cpdf[['Year', 'Scaling factor']], on='Year', how='left')
df2['Adjusted Pt6 time'] = df2['Pt6 time'] * df2['Scaling factor']


plt.figure(figsize=(10, 6))

# Plotting KDE for adjusted finish times for each year.
years = df2['Year'].unique()
for year in sorted(years):
    sns.kdeplot(
        df2[df2['Year'] == year]['Adjusted Pt6 time'],
        label=str(year),
        fill=False,  # Remove fill for better distinction between curves
        common_norm=False,  # Don't normalize across years
        warn_singular=False
    )

# Customizing the x-axis to show HH:MM format
tick_positions = np.arange(240, 1200, 60) 
plt.xticks(
    ticks=tick_positions, # Labels on the x-axis every 30 minutes
    labels=[f"{int(m//60):02d}:{int(m%60):02d}" for m in tick_positions],  # HH:MM labels
    rotation=45
)

# Customizing the plot.
plt.title('Adjusted Finish Times Distribution by Year', fontsize=16)
plt.xlabel('Adjusted Finish Time', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend(title='Year')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Modeling

For the modeling part, we'll investigate the impact of positive elevation for different groups of runners.

The basic assumption is that 100m of elevation = 1km flat, but that's probably not true to the same degree for all runners. We've seen in the previous part that elite participants (eg, the top 10) seemed to not be very affected by year to year changes in conditions. Since we have race results with a few different distances and a few different elevations, we can try to fit a simple linear regression model on those variables :

$finishTime = a*distance + b*elevation + c$

That way, if our model learns well, we can estimate the impact of added elevation from one year's Saint√©Lyon to another, compared to the impact of added distance, and this for different groups of runners.

```python
from sklearn.linear_model import LinearRegression

df3 = df2.merge(cpdf[['Year', 'Pt6 distance', 'Pt6 total elevation']], on='Year', how='left')

X = df3[['Pt6 distance', 'Pt6 total elevation']].values  # Features
y = df3['Pt6 time'].values  # Target

# Linear regression
model = LinearRegression(positive=True)
model.fit(X, y)

# Coefficients of the model : getting "a" and "b"
distance_coef = model.coef_[0]  # Impact of 1km flat on time
elevation_coef = model.coef_[1]  # Impact of 1m of elevation on time

# Equivalence between elevation and distance
elevation_to_distance = 100*elevation_coef / distance_coef  # in km flat per m of elevation

print("For all runners on average :")
print(f"Impact of 1km flat of distance added (in minutes) : {distance_coef:.2f}")
print(f"Impact of 100m of elevation added (in minutes) : {elevation_coef*100:.2f}")
print(f"=> 100m of elevation corresponds to {elevation_to_distance:.2f} km.")

r2 = model.score(X, y)
print(f"R¬≤ of the model : {r2:.2f}")
```

We can see that for such a big group, we have a low R¬≤ score : there is too much variability in the finish times to be explained correctly only by these two values.

So, let's try to form some gradually smaller groups of runners : the top 1000, the top 100, the top 10, and the 100 runners closest to the average for their year.  We'll still try to group by sex to see what happens.

```python
### Creating the groups

average_100_runners = []
for year in df3['Year'].unique():
    year_data = df3[df3['Year'] == year]
    center_rank = year_data['Pt6 rank'].max() // 2 # "centered" rank (half of the last rank)
    # Select the runners from center_rank - 50 to center_rank + 50
    selected_year_data = year_data[(year_data['Pt6 rank'] >= center_rank - 50) & 
                                   (year_data['Pt6 rank'] <= center_rank + 50)]
    average_100_runners.append(selected_year_data)
average_100_runners_df = pd.concat(average_100_runners).reset_index(drop=True)

lowest_100_runners = []
for year in df3['Year'].unique():
    year_data = df3[df3['Year'] == year]
    last_rank = year_data['Pt6 rank'].max()
    # Select the runners from last rank to last rank -100
    selected_year_data = year_data[(year_data['Pt6 rank'] >= last_rank - 100)]
    lowest_100_runners.append(selected_year_data)
lowest_100_runners_df = pd.concat(lowest_100_runners).reset_index(drop=True)

top_10_mask = df3['Pt6 rank'] <= 10
top_100_mask = df3['Pt6 rank'] <= 100
top_1000_mask = df3['Pt6 rank'] <= 1000
women_mask = df3['Sex'] == 'F'
men_mask = df3['Sex'] == 'H'

groups = {"Men":df3[men_mask], "Women":df3[women_mask], 
          "Top 1000":df3[top_1000_mask], "Top 100":df3[top_100_mask], "Top 10":df3[top_10_mask],
          "100 average ranked":average_100_runners_df, "100 lowest ranked":lowest_100_runners_df}
```

```python
results = []

for name, subdf in groups.items():
    x = subdf[['Pt6 distance', 'Pt6 total elevation']].values
    y = subdf['Pt6 time'].values 

    model = LinearRegression(positive=True)
    model.fit(x, y)
    
    distance_coef = model.coef_[0]
    elevation_coef = model.coef_[1]
    elevation_to_distance = 100 * elevation_coef / distance_coef
    r2 = model.score(x, y)
    
    # Store results in a list
    results.append({
        "Group": name,
        "Impact of 1km Distance (min)": f"{distance_coef:.2f}",
        "Impact of 100m Elevation (min)": f"{elevation_coef * 100:.2f}",
        "100m Elevation in km": f"{elevation_to_distance:.2f}",
        "R¬≤": f"{r2:.2f}"
    })

# Convert results to a DataFrame and display
results_df = pd.DataFrame(results)
results_df.head(10)
```

We can see that for groups with a lower result variability, such as the last 3, the R¬≤ score indicates that the model is explaining the variability quite well. It's normal to not reach a very high R¬≤ with only 2 simple variables like that : we can't fully explain results variability without factors like weather, course technicity, etc...

What's striking here is the difference of impact the elevation has on the elite vs lower ranked : basically the stronger trail runner you are, the less you are impacted by elevation, up to almost not being impacted at all ! This might be explained by the fact that elite trail runners, who are already very fast in uphill sections, can compensate very well for the lost time by going much, much faster than average runners in downhill sections, thanks to their skill.

On the contrary, the lowest ranked runners of each race seem to be affected more by elevation than by distance itself.

For the average, we can see that the famous 100m d+ = 1km flat is not too far fetched : we get an estimation of 0.74km per 100m. The group of runners for which the assumption is most true must be between the average and the lower ranks.




Let's visualise what our model is telling us by making it predict times with a fixed distance or fixed elevation, and a varying 2nd variable. We'll do that on the 3 groups the model fitted best to :

```python
### Predicted finish times vs elevation and vs distance

groups = {"Top 10": df3[top_10_mask], "100 average ranked": average_100_runners_df, "100 lowest ranked": lowest_100_runners_df}

# Fixed distance, varying elevation
fixed_distance = 80
elevations = np.linspace(1000, 4000, 100)
pred_times_distance = []
for name, subdf in groups.items(): 
    x = subdf[['Pt6 distance', 'Pt6 total elevation']].values
    y = subdf['Pt6 time'].values
    model = LinearRegression(positive=True)
    model.fit(x, y)

    x_test = np.array([[fixed_distance, elev] for elev in elevations])
    pred_times_distance.append(model.predict(x_test))

# Fixed elevation, varying distance
fixed_elevation = 2000
distances = np.linspace(60, 90, 100)
pred_times_elev = []
for name, subdf in groups.items(): 
    x = subdf[['Pt6 distance', 'Pt6 total elevation']].values
    y = subdf['Pt6 time'].values
    model = LinearRegression(positive=True)
    model.fit(x, y)

    x_test = np.array([[distance, fixed_elevation] for distance in distances])
    pred_times_elev.append(model.predict(x_test))


# Plotting
colors = ['blue', 'green', 'red']
labels = ['Top 10', '100 average ranked', '100 lowest ranked']
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot for fixed distance vs elevations
for predicted, color, label in zip(pred_times_distance, colors, labels):
    axs[0].plot(elevations, predicted, color=color, label=label)
axs[0].set_xlabel('Elevation (m)')
axs[0].set_ylabel('Predicted Time (minutes)')
axs[0].set_title('Predicted times vs elevation (fixed distance)')
axs[0].legend()
axs[0].grid(True)

# Plot for fixed elevation vs distances
for predicted, color, label in zip(pred_times_elev, colors, labels):
    axs[1].plot(distances, predicted, color=color, label=label)
axs[1].set_xlabel('Distance (km)')
axs[1].set_ylabel('Predicted Time (minutes)')
axs[1].set_title('Predicted times vs distance (fixed elevation)')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

We clearly see here that what affects your performance is very different depending on your level : the better you are, the more the distance is the main factor, and inversely. Of course, this is only part of the whole picture and is probably very dependent on negative elevation as well, but the difference is still impressive.

Basically, you can tire out average runners by adding more elevation, but for the elite... They'll have to go down and they'll catch up eventually.

What are the main factors explaining performances ? We can't say that we've explored them all here, but we've made an important observation : these factors vary greatly depending on the athlete, his skill level and his running habits. For a race like the Saint√©Lyon, we could certainly explain performance very accurately by deep diving into meteorological factors impacting the terrain, the pedigree of the runners, and the distance and elevation.
